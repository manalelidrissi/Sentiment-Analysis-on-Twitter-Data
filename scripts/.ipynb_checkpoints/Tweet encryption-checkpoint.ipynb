The purpose of this Lab is to encrypt and decrypt tweets in data previously collected from Twitter.

### Objectives of this jupyter notebooks file:
- Tweet Encryption and decryption

#### Dependencies import
 

```python
import pandas as pd
from cryptography.fernet import Fernet
```

## Load Datasets

```python
df_usa = pd.read_csv("../cleaned_data/usa_cleaned_tweets.csv")
df_uk = pd.read_csv("../cleaned_data/uk_cleaned_tweets.csv")
df_germany = pd.read_csv("../cleaned_data/germany_cleaned_tweets.csv")
```

## Tweet Encryption and Decryption

```python
def encrypt(tweet):
    """
    Encrypts a tweet
    """
    key = Fernet.generate_key()
    encoded_tweet = tweet.encode()
    f = Fernet(key)
    encrypted_tweet = f.encrypt(encoded_tweet)

    return pd.Series([key, encrypted_tweet])

def decrypt(key, encrypted_tweet):
    """
    Decrypts an encrypted tweet
    """
    f = Fernet(key)
    decrypted_tweet = f.decrypt(encrypted_tweet).decode()

    return decrypted_tweet
```

Let's take the df_usa as an example

```python
df_usa[["key","encrypted_tweet"]] = df_usa["cleaned_tweet"].apply(encrypt)
df_usa["decrypted_tweet"] = df_usa.apply(lambda row: decrypt(row["key"], row["encrypted_tweet"]), axis=1)
```
